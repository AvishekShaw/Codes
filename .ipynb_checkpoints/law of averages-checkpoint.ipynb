```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
up_bound = 10000
low_bound = 0
step_size = 100
```

```python
num_tosses_arr = np.arange(low_bound,up_bound,step_size)
num_heads_list = []
```

```python
for num_tosses in num_tosses_arr:
    num_heads = 0
    for i in range(num_tosses):
        toss = np.random.randint(0,2)
        if toss == 1:
            num_heads+=1
    num_heads_list.append(num_heads)
```

```python
num_heads_arr = np.array(num_heads_list)
chance_error_arr = num_tosses_arr/2 - num_heads_arr 
```

```python
fig,ax = plt.subplots(figsize=(20,10))
ax.scatter(num_tosses_arr, num_heads_arr,label="num heads",color='red')
ax.scatter(num_tosses_arr,chance_error_arr,label="chance error",color="blue")
ax.plot(num_tosses_arr, num_tosses_arr/2,label="y=x/2")
ax.set(xlim=(low_bound,up_bound),ylim=(-100,up_bound/2))
ax.legend()
```
